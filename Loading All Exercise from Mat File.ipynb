# Importing Libraires 

```python
import pandas as pd
from scipy.io import loadmat
```

# Loading Exercise 1

```python
Exercise_1 = loadmat(r"S1_E1_A1.mat")
```

```python
print(Exercise_1)
```

```python
print(Exercise_1.keys())
```

```python
emg_1 = Exercise_1['emg']
stimulus_1 = Exercise_1['stimulus']
restimulus_1 = Exercise_1['restimulus']
repetition_1 = Exercise_1['repetition']
rerepetition_1 = Exercise_1['rerepetition']
```

# DataFramae 

### Convert the dictionary into DataFrame

```python
# Custom Column Names 
```

```python
df_1= pd.DataFrame(emg_1, 
                  columns = ['Emg1','Emg2','Emg3','Emg4','Emg5','Emg6','Emg7','Emg8','Emg9','Emg10','Emg11','Emg12'])
```

```python
df_1
```

```python
repetition_1 = pd.DataFrame(repetition_1)
rerepetition_1 = pd.DataFrame(rerepetition_1)
stimulus_1 = pd.DataFrame(stimulus_1)
restimulus_1 = pd.DataFrame(restimulus_1)
```

### Adding new column to existing DataFrame

```python
df_1['repetition'] = repetition_1
df_1['rerepetition'] = rerepetition_1
df_1['stimulus'] = stimulus_1
df_1['restimulus'] = restimulus_1
```

# Final Output of Exercise 1 DataFrame

```python
df_1
```

```python
print(df_1.describe(include = 'all'))
print("************************************************************************************************************")
print(df_1.info())
print("************************************************************************************************************")
print("Shape of the Dataset is", df_1.shape)
```
