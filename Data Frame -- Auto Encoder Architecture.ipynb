# Auto Encoder Architecture

## Importing Libraries

```python
from tensorflow.keras.utils import plot_model
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input
from tensorflow.keras.layers import Dense
from keras.models import Sequential
from keras.layers import Dense, BatchNormalization, Dropout
from tensorflow.keras.optimizers import Adam, SGD, RMSprop, Adadelta, Adagrad, Adamax, Nadam, Ftrl
```

# 1. Define Network

```python
n_inputs = X.shape[1]
```

```python
# define encoder
visible = Input(shape=(n_inputs,))


# encoder level 1
e = Dense(n_inputs*2, activation = 'relu')(visible)




# encoder level 2
e = Dense(n_inputs, activation = 'softmax')(e)




# bottleneck
n_bottleneck = 3
bottleneck = Dense(n_bottleneck)(e)



# define decoder, level 1
d = Dense(n_inputs, activation ='softmax')(bottleneck)



# decoder level 2
d = Dense(n_inputs*2, activation = 'relu')(d)






# output layer
output = Dense(n_inputs, activation='softmax')(d)

```

```python
# define autoencoder model
model = Model(inputs=visible, outputs=output)
```

# 2. Compile Network

```python
# compile autoencoder model
model.compile(optimizer='adam', loss='mse', metrics=['accuracy'])
```

```python
# plot the autoencoder
plot_model(model, 'autoencoder_no_compress.png', show_shapes=True)
```

```python
model.summary()
```

# 3. Fit Network

```python
# fit the autoencoder model to reconstruct input
history = model.fit(X_train, X_train, epochs=1, batch_size=64, validation_data=(X_test,X_test))
```

# 4.Evaluate Network

## 4.1. Plotting Loss and Accuracy

```python
#Loss Curve


plt.plot(figsize = [8,6])

plt.plot(history.history['loss'],'r',linewidth = 3.0)
plt.plot(history.history['val_loss'],'b',linewidth = 3.0)

plt.legend(['Training Loss', 'Validation Loss'], fontsize = 18)

plt.xlabel('Epochs', fontsize = 16)
plt.ylabel('Loss', fontsize = 16)

plt.title('Loss Curve of Deep Neural Netowork' , fontsize = 16 )

```

```python
#Accuracy Curve

plt.plot(figsize = [8,6])
plt.plot(history.history['accuracy'],'r',linewidth = 3.0)
plt.plot(history.history['val_accuracy'],'b',linewidth = 3.0)

plt.legend(['Training Accuracy', 'Validation Accuracy'], fontsize = 18)

plt.xlabel('Epochs', fontsize = 16)
plt.ylabel('Accuracy', fontsize = 16)


plt.title('Accuracy Curve of Deep Neural Netowork' , fontsize = 16 )
```
